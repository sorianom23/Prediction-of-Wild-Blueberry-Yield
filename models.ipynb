## Training Models

### Linear Regression 

Score = 0.81 on both

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, 1:-1]
y = df[['yield']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
print("lr.coef_: {}".format(lr.coef_))
print("lr.intercept_: {}".format(lr.intercept_))
```

```python
print("Training set score: {:.2f}".format(lr.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lr.score(X_test, y_test)))
```

### Ridge Regression 

L2 Regularization

```python
from sklearn.linear_model import Ridge
```

```python
X = df.iloc[:, [-2, -3, -4]]
y = df[["yield"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state=44)
```

```python
ridge = Ridge().fit(X_train, y_train)
```

```python
print("Training set score: {:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge.score(X_test, y_test)))
```

##### Restricting more 

```python
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge10.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge10.score(X_test, y_test)))
```

##### Restricting less 

```python
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
print("Training set score: {:.2f}".format(ridge01.score(X_train, y_train)))
print("Test set score: {:.2f}".format(ridge01.score(X_test, y_test)))
```

### K-NNeighbor

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
```

```python
X = df.iloc[:, 1:-1]
y = df[['yield']]
```

```python
#Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
#Normalize
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsRegressor().fit(X_train_scaled, y_train)
```

```python
print("Train set R^2: {:.2f}".format(knn.score(X_train_scaled, y_train)))
print("Test set R^2: {:.2f}".format(knn.score(X_test_scaled, y_test)))
```

#### Neighbors n=20

```python
knn20 = KNeighborsRegressor(n_neighbors = 20).fit(X_train_scaled, y_train)
```

```python
print("Train set R^2: {:.2f}".format(knn20.score(X_train_scaled, y_train)))
print("Test set R^2: {:.2f}".format(knn20.score(X_test_scaled, y_test)))
```

### GradientBoosting

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbrt0 = GradientBoostingRegressor(random_state = 0)

gbrt0.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(gbrt0.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt0.score(X_test, y_test)))
```

#### limiting max_depth

```python
gbrt1 = GradientBoostingRegressor(random_state = 0, max_depth=4)

gbrt1.fit(X_train, y_train)

print("Accuracy on training set: {:.3f}".format(gbrt1.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(gbrt1.score(X_test, y_test)))
```

#### limiting learning rate
